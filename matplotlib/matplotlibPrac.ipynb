```python
plt.plot([1,3,6,9,13])
plt.show()
```

```python
plt.plot([3,4,5,6,7,8], [13,16,21,25,29,44], 'ro-') # x,y 축 및 빨간선
plt.show()
```

```python
plt.plot([3,4,5,6,7,8], [13,16,21,25,29,44], 'ro-')
plt.axis([2,9,10,40]) # x축과 y축의 범위를 지정
plt.show()
```

```python
plt.plot([3,4,5,6,7,8], [13,16,21,25,29,44], 'ro-')
plt.axis([2,9,10,40])
plt.xlabel("month") # 라벨 적기
plt.ylabel("temperature")
plt.show()
```

```python
# y=x, y=x**2, y=x**3 그래프 그리기
import numpy as np
```

```python
x_range = np.arange(0,5,0.2)
```

```python
x_range
```

```python
plt.plot(x_range, x_range, 'b--', x_range, x_range**2, 'ys-', x_range,x_range**3, 'g^-') # 파란색 점선 / 노란색 선 / 초록색 세모선
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph')
plt.show()
```

```python
# '색 r,b,y,g
# - 1개면 실선, -- 2개면 점선
# ^ 세모
# s 네모
# o 동그라미
```
