```python
## 각 '구' 별로 전체 인구대비 고령자가 얼마나 많은지 확인해보자
```

```python
# 필요한 칼럼 가져오기
report2 = report.filter(items=['자치구', '동', '인구', '65세이상고령자'])
```

```python
report2
```

```python
# 각 구별 전체 인원만 뽑아내면서 새로운 데이터 프레임 생성
report3 = DataFrame(list(map(lambda x : x if x[1]=='소계' else () , report2.values[3:])), columns=['Gu', 'Dong', 'Total', 'Older'])
```

```python
report3
```

```python
# none 으로 채워진 row 삭제
report4= report3.dropna()
```

```python
report4
```

```python
report5 = report4.filter(items=['Gu', 'Total', 'Older']) # 동(소계) 즉, 필요없는 데이터 삭제
```

```python
report5
```

```python
# 데이터를 받아서 float 형태로 변경(, 제거)하는 함수
def float_replace(x):
    return float(x.replace(',',''))
```

```python
# 데이터를 숫자로 변경
report6 = DataFrame(list(map(lambda x: [ x[0], float_replace(x[1]), float_replace(x[2]) ], report5.values)), columns=report5.columns)
```

```python
report6
```

```python
# 인덱스를 구별로 조작
report7 = DataFrame(report6.filter(items=['Total','Older']).values, index=report6.Gu.values, columns = report6.columns[1:])
```

```python
report7
```

```python
# 구별로 그래프 그리기 (x축은 그래프임... 폰트를 한글 폰트 지원으로 바꾸면 됨)
report7.plot(kind='bar')
plt.show()
```
