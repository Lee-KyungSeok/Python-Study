```python
import pandas as pd
from pandas import Series
from matplotlib import pyplot as plt
```

```python
# csv 파일을 읽어서 -> dataframe의 형태로 변환시켜줌
## 주의 : csv 는 ',(콤마)'로 분리되어 있지만 아래는 탭으로 분리되어 있으므로 이를 변환시켜 줘야 한다.
# pd.read_csv('./report.txt') # 사용 불가
```

```python
# 따라서 table을 읽을 수 있는 기능을 이용
report = pd.read_table('./report.txt')
```

```python
report
```

```python
report.ix[2]
```

```python
report.ix[2].values
```

```python
## 필요한 데이터만 Series로 만든다.
## 타입은 float형으로 변환
## 필요한 데이터는 3번째 column 부터이므로 이를 가져옴
seoul_people = Series(map(lambda x : float(x.replace(',','')), report.ix[2].values[3:]), 
       index=map(lambda x: "{main} ({sub})".format(main=x[0], sub=x[1]), 
                 zip(report.ix[0].values[3:], report.ix[1].values[3:])))
```

```python
seoul_people
```

```python
seoul_people.plot(kind='bar')
plt.show()
```

```python
fig = plt.figure()
```

```python
# 네가지의 그래프 틀 만들기 ()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```
